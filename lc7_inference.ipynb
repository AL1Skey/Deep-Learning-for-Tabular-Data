# Import Library

```python
import pandas as pd
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np
```

# Data Loading

```python
data=pd.read_csv('inference.csv')
data.info()
```

```python
model = keras.models.load_model('func_model.hdf5')
with open('feature.json','r') as f:
    import json
    feature = json.load(f)
import joblib
trans = joblib.load('transformer.pkl')

trained = data[feature]
trained = trans.transform(trained)
pred = np.where(model.predict(trained)>0.5,'Cancelled','Not Canceled')
```

```python
data['pred'] = pred
data
```
